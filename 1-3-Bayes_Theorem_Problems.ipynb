# Bayes Theorem in Python

This notebook walks through Bayes' Theorem with examples and interactive components.

## 1. Introduction to Bayes' Theorem
Bayes' Theorem allows us to update our beliefs based on new evidence.

**Formula:**
\[ P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} \]

```python
# Medical Test Example
print("Medical Test Example:")

# Prior probability of having the disease
p_disease = 0.01  # 1%

# Probability of testing positive given disease (sensitivity)
p_pos_given_disease = 0.99

# Probability of testing positive given no disease (false positive rate)
specificity = 0.95
p_pos_given_no_disease = 1 - specificity

# Probability of not having the disease
p_no_disease = 1 - p_disease

# Total probability of testing positive
p_positive = (p_pos_given_disease * p_disease) + (p_pos_given_no_disease * p_no_disease)

# Applying Bayes' Theorem
p_disease_given_positive = (p_pos_given_disease * p_disease) / p_positive
print(f"Probability of having the disease given a positive test result: {p_disease_given_positive:.2%}\n")
```

```python
# Spam Classifier Example
print("Spam Classifier Example:")

# Prior probability that an email is spam
p_spam = 0.2

# Probability that a spam email contains the word "win"
p_win_given_spam = 0.7

# Probability that a non-spam email contains "win"
p_win_given_ham = 0.01

# Probability of not being spam
p_ham = 1 - p_spam

# Total probability of seeing the word "win"
p_win = (p_win_given_spam * p_spam) + (p_win_given_ham * p_ham)

# Applying Bayes' Theorem
p_spam_given_win = (p_win_given_spam * p_spam) / p_win
print(f"Probability that an email is spam given it contains 'win': {p_spam_given_win:.2%}\n")
```

```python
# Interactive Bayes Calculator
try:
    from ipywidgets import interact, FloatSlider

    def bayes_calculator(p_a=0.5, p_b_given_a=0.5, p_b_given_not_a=0.5):
        p_not_a = 1 - p_a
        p_b = (p_b_given_a * p_a) + (p_b_given_not_a * p_not_a)
        p_a_given_b = (p_b_given_a * p_a) / p_b if p_b != 0 else 0
        print(f"P(A|B) = {p_a_given_b:.2%}")

    interact(
        bayes_calculator,
        p_a=FloatSlider(value=0.5, min=0.01, max=1.0, step=0.01, description='P(A)'),
        p_b_given_a=FloatSlider(value=0.5, min=0.01, max=1.0, step=0.01, description='P(B|A)'),
        p_b_given_not_a=FloatSlider(value=0.5, min=0.01, max=1.0, step=0.01, description='P(B|¬A)')
    )

except ModuleNotFoundError:
    print("ipywidgets not installed. Run `pip install ipywidgets` to enable interactive demo.")
```
