#  Exploratory Data Analysis 





 **John Tukey, a preeminent statistician, whose ideas developed over fifty years ago form the foundation of data science.**
 
            Statistics emerged as a modern discipline in the past century, built upon probability theory developed between the 17th and 19th centuries by thinkers like Bayes, Laplace, and Gauss. While probability remains a theoretical framework, statistics evolved into an applied science focused on analyzing and modeling data. Key figures such as Galton, Pearson, and Fisher laid the groundwork for modern statistical methods like experimental design and maximum likelihood estimation.

    A significant shift occurred in 1962 when John W. Tukey, in his influential paper “The Future of Data Analysis,” proposed a broader vision for the field—emphasizing data analysis as a distinct scientific discipline. He introduced the concept of exploratory data analysis (EDA), advocating for its role in uncovering patterns beyond traditional inference. Tukey’s ideas, strengthened by advancements in computing and software, have become foundational to modern data science. Today, EDA continues to grow, driven by technological progress, access to large datasets, and widespread use across disciplines.

# 2.  Estimates of Location – Summary

When analyzing numerical data, we often want to know where the values are "centered." These are called **estimates of location**.

---

###  Mean (Average)

The mean is the sum of all values divided by the number of values:

$$
\bar{x} = \frac{1}{N} \sum_{i=1}^{N} x_i
$$

---

###  Trimmed Mean

This is the mean after removing \( p \) smallest and \( p \) largest values from the sorted dataset:

$$
\text{Trimmed Mean} = \frac{1}{N - 2p} \sum_{i=p+1}^{N-p} x_i
$$

Used to reduce the impact of outliers.

---

###  Weighted Mean

Each value $( x_i )$ is assigned a weight $( w_i )$:

$$
\bar{x}_w = \frac{\sum_{i=1}^{N} w_i x_i}{\sum_{i=1}^{N} w_i}
$$

Useful when some values are more important or reliable than others.

---

###  Median

The median is the middle value in a **sorted** dataset. If there’s an even number of values, take the average of the two middle values.  
It is also known as the **50th percentile**, and is **robust to outliers**.

---

###  Weighted Median

The weighted median is the point in the sorted data where the cumulative weight reaches 50%. Like the median, it's robust.

---

###  Outliers and Robust Estimates

- **Outlier**: A value far from the rest of the data.
- **Robust Estimate**: Not sensitive to extreme values (e.g., median, trimmed mean).
- **Mean** is sensitive to outliers.
- **Median** and **trimmed mean** offer better protection in the presence of anomalies.

---

###  Key Insight

Use **mean** when the data is clean and symmetric.  
Use **median** or **trimmed mean** when the data contains **outliers** or is **skewed**.


## 3. Estimates of Variability
Location is one aspect of summarizing a feature, and variability (or dispersion) is the second. Variability measures whether data values are tightly clustered or widely spread out.
### 3.1 Mean Absolute Deviation (MAD)

The **Mean Absolute Deviation (MAD)** is calculated as:

$$
\text{Mean Absolute Deviation (MAD)} = \frac{1}{N} \sum_{i=1}^{N} \left| x_i - \bar{x} \right|
$$

Where:
- $( x_i $) is the data point,
- $( \bar{x} $) is the sample mean,
- $( N  $) is the number of data points.

### 3.2 Variance

The **Variance** is given by:

$$
\text{Variance} (s^2) = \frac{1}{N-1} \sum_{i=1}^{N} (x_i - \bar{x})^2
$$

Where:
- $( x_i $) is each data point,
- $( \bar{x} $) is the sample mean,
- $( N $) is the number of data points.

### 3.3 Standard Deviation

The **Standard Deviation** is the square root of the variance:

$$
\text{Standard Deviation} (s) = \sqrt{\frac{1}{N-1} \sum_{i=1}^{N} (x_i - \bar{x})^2}
$$

### 3.4 Median Absolute Deviation (MAD from Median)

The **Median Absolute Deviation** (MAD) is computed as:

$$
\text{Median Absolute Deviation (MAD)} = \text{Median} \left( \left| x_1 - m \right|, \left| x_2 - m \right|, \ldots, \left| x_N - m \right| \right)
$$

Where:
- \( m \) is the median of the data.

### 3.5 Percentile Definition

For a **P-th percentile**, the formula can be expressed as:

$$
\text{Percentile}_P = (1 - w) \cdot x_j + w \cdot x_{j+1}
$$

Where:
- $( x_j $) and $( x_{j+1} $) are the order statistics (sorted data points),
- $( w $) is a weight between 0 and 1.

### 3.6 Interquartile Range (IQR)

The **Interquartile Range (IQR)** is defined as:

$$
\text{IQR} = \text{Percentile}_{75} - \text{Percentile}_{25}
$$

This is the difference between the 75th percentile and the 25th percentile.


### 3. Exploring the Data Distribution

While location and variability summarize data with single values, it's also important to explore how the **data is distributed overall**.

---

####  Key Terms for Exploring the Distribution

- **Boxplot (Box and Whiskers Plot):**  
  A graphical summary introduced by Tukey to visualize the distribution, center, and spread of the data.

- **Frequency Table:**  
  A count of numeric values falling into specific intervals (bins).

- **Histogram:**  
  A bar chart of the frequency table, showing bins on the x-axis and counts (or proportions) on the y-axis.

- **Density Plot:**  
  A smooth version of the histogram, often created using kernel density estimation (KDE).

---

#### Statistical Moments

In statistics, the distribution's shape is described using **moments**:

- **1st Moment – Mean (Location):** Where the data is centered.  
- **2nd Moment – Variance (Variability):** How spread out the data is.  
- **3rd Moment – Skewness:** Indicates if the data leans towards higher or lower values.  
- **4th Moment – Kurtosis:** Measures the likelihood of extreme values (heavy tails).
